```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")

print(f"El archivo tiene {df.shape[0]} filas y {df.shape[1]} columnas.")

df.head(3)

```

```python
valores_nulos = df.isnull().sum()

print(valores_nulos)
```

```python
df["director"] = df["director"].fillna("Director desconocido")
df["cast"] = df["cast"].fillna("Elenco no disponible")
df["country"] = df["country"].fillna("Pais no especificado")

print(df.isnull().sum())

```

```python
df = df.dropna(subset=["date_added", "rating", "duration"])

df = df.drop(columns=["show_id", "description"])

print(df.isnull().sum())
print(f"\nTamaño actual del set de datos: {df.shape}")
```

```python
df["release_year"] = df["release_year"].astype(int)

print(df["release_year"].head(3))

```

```python
df["genero_principal"] = df["listed_in"].str.split(",").str[0]

df[["listed_in","genero_principal"]].head(5)
```

```python
df.to_csv("netflix_titles_limpio.csv", index=False)
print("Archivo guardado con exito")
```
